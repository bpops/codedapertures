# CAPPY Examples

CAPPY (Coded Aperture Production in PYthon) is a python module that allows users to quickly and easily generate common coded aperture patterns.

A URA may be generated by providing the rank, $l$, which is the zero-indexed $l$th pair of prime numbers, $r$ and $s$, that satisfy the criteria $r-s=2$. The multiplier is how many times the pattern is repeated. The pattern is automatically centered.

```python
import cappy
ura = cappy.ura(rank=2, mult=5)
```

The mask can be displayed.

```python
ura.show()
```

Or, the pattern can be provided in a numpy array.

```python
pattern = ura.get_pattern()
print(pattern.shape)
```

Similarly, a random array may be generated with a specified $r$ and $s$, and fill factor.

```python
rand_mask = cappy.rand_array(r=100, s=80, fill=0.6)
rand_mask.show()
```
