# CAPPY Examples

CAPPY (Coded Aperture Production in PYthon) is a python module that allows users to quickly and easily generate common coded aperture patterns.

Below, we show examples of building and showing various coded aperture patterns.

```python
import codedapertures as cappy
```

### Uniformly Redundant Array

```python
ura = cappy.ura(rank=2, mult=5)
ura.add_border(2)
ura.show()
```

### Modified Uniformly Redundant Array

```python
mura = cappy.mura(rank=11, mult=1)
mura.show(inverse=True)
```

### Random Array

```python
rand_mask = cappy.rand_array(r=100, s=80, fill=0.6)
rand_mask.show()
```
